```python
############Matplotlib
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Line plot
t=np.arange(0.0,2.0,0.01)
s=1+np.sin(2*np.pi*t)
fig,ax=plt.subplots() #fig includes ax
ax.plot(t,s)
ax.set(xlabel='time(s)' ,ylabel='voltage', title='About as simple as it gets, folks')
ax.grid()

fig.savefig('test.png')
plt.show()
```

```python
#Multiple subplots in one figure
#multiple axes are created with subplots func
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.title('  ')
plt.ylabel('Demped oscillation')

plt.subplot(2,1,2)
plt.plot(x2,y2)
```
